# Smoothing Techniques in Time Series
Smoothing techniques help remove noise from time series data to reveal underlying trends and patterns. The two most common techniques are:

1. **Moving Averages (MA)**
2. **Exponential Smoothing (ES)**
- A moving average smooths fluctuations by taking the average of past values over a fixed window.

# 1. Types of Moving Averages
- Simple Moving Average (SMA)
- Weighted Moving Average (WMA)

## (a)**Simple Moving Average (SMA)**

        The Simple Moving Average (SMA) smooths fluctuations by averaging past values.

$$
SMA_t = \frac{1}{N} \sum_{i=t-N+1}^{t} X_i
$$

where:
- \( SMA_t \) is the moving average at time \( t \),
- \( X_i \) are the past observations,
- \( N \) is the window size.

A larger \( N \) results in a smoother curve but increases lag.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample stock price data
data = {'Date': pd.date_range(start='2024-01-01', periods=50, freq='D'),
        'Close': np.random.randint(100, 200, size=50)}
stock_data = pd.DataFrame(data)
stock_data.set_index('Date', inplace=True)

# Define SMA function
def calculate_sma(data, window):
    return data.rolling(window=window).mean()

# Calculate 10-day and 20-day SMA
stock_data['SMA_10'] = calculate_sma(stock_data['Close'], window=10)
stock_data['SMA_20'] = calculate_sma(stock_data['Close'], window=20)

# Plotting the SMA
plt.figure(figsize=(12,6))
plt.plot(stock_data['Close'], label='Stock Price', color='blue', alpha=0.5)
plt.plot(stock_data['SMA_10'], label='10-Day SMA', color='orange', linewidth=2)
plt.plot(stock_data['SMA_20'], label='20-Day SMA', color='red', linewidth=2)

plt.title('Simple Moving Average (SMA)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

## (b)**Weighted Moving Average (WMA)**

The **Weighted Moving Average (WMA)** is a type of moving average where more recent data points are given higher weights than older data points. Unlike the Simple Moving Average (SMA), which assigns equal weight to all values, WMA assigns different weights to each value, making it more responsive to recent price changes.

#### **Formula for WMA:**

$$
WMA_t = \frac{\sum_{i=0}^{N-1} w_i X_{t-i}}{\sum_{i=0}^{N-1} w_i}
$$

where:
- \( WMA_t \) is the weighted moving average at time \( t \),
- \( X_{t-i} \) are the past observations,
- \( w_i \) are the assigned weights (typically decreasing as we go back in time),
- \( N \) is the window size.

#### **Example Weights:**
For a 3-day WMA, you might assign weights as:
- Day 1: **1**
- Day 2: **2**
- Day 3: **3**  
(Summing to **6**: \(1+2+3 = 6\))

#### **Key Characteristics:**
- More recent data has a **higher impact** on the moving average.
- Reacts **faster** to price changes compared to SMA.
- Suitable for detecting short-term trends.

#### **Comparison with SMA:**
| Feature | SMA | WMA |
|---------|-----|-----|
| Weighting | Equal | Higher for recent values |
| Sensitivity | Lower | Higher |
| Use case | Long-term trends | Short-term trends |

By adjusting the weights, WMA can be fine-tuned to capture trends more effectively.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample stock price data
data = {'Date': pd.date_range(start='2024-01-01', periods=50, freq='D'),
        'Close': np.random.randint(100, 200, size=50)}
stock_data = pd.DataFrame(data)
stock_data.set_index('Date', inplace=True)

# Function to calculate Weighted Moving Average (WMA)
def calculate_wma(data, window):
    weights = np.arange(1, window + 1)  # Weights: 1, 2, 3, ..., window
    return data.rolling(window).apply(lambda x: np.dot(x, weights) / weights.sum(), raw=True)

# Calculate 10-day and 20-day WMA
stock_data['WMA_10'] = calculate_wma(stock_data['Close'], window=10)
stock_data['WMA_20'] = calculate_wma(stock_data['Close'], window=20)

# Plotting the WMA
plt.figure(figsize=(12,6))
plt.plot(stock_data['Close'], label='Stock Price', color='blue', alpha=0.5)
plt.plot(stock_data['WMA_10'], label='10-Day WMA', color='orange', linewidth=2)
plt.plot(stock_data['WMA_20'], label='20-Day WMA', color='red', linewidth=2)

plt.title('Weighted Moving Average (WMA)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

### 2.**Exponential Moving Average (EMA)**

The **Exponential Moving Average (EMA)** is a type of moving average that places more weight on recent data points, making it more responsive to new information compared to the Simple Moving Average (SMA). It is widely used in time series analysis and financial markets.

#### **Formula for EMA:**

The EMA is calculated recursively using the following formula:

$$
EMA_t = \alpha X_t + (1 - \alpha) EMA_{t-1}
$$

where:
- \( EMA_t \) = Exponential Moving Average at time \( t \),
- \( X_t \) = Current value at time \( t \),
- \( \alpha \) = Smoothing factor (between 0 and 1),
- \( EMA_{t-1} \) = Previous EMA value.

#### **Smoothing Factor \( alpha ):**
The smoothing factor \( alpha ) is calculated as:

$$
\alpha = \frac{2}{N+1}
$$

where:
- \( N \) = Number of periods in the EMA.

#### **Key Characteristics:**
- **More weight on recent data**  Reacts quickly to changes.
- **Less lag compared to SMA**  Useful for trend-following strategies.
- **Commonly used in finance**  Helps in technical analysis (e.g., crossover strategies).

#### **Comparison: EMA vs. SMA**
| Feature | SMA | EMA |
|---------|-----|-----|
| Weighting | Equal | Higher for recent values |
| Sensitivity | Lower | Higher |
| Lag | More | Less |
| Use Case | Long-term trends | Short-term trends |

#### **Example:**
For a **10-day EMA**, the smoothing factor would be:

$$
\alpha = \frac{2}{10+1} = \frac{2}{11} \approx 0.1818
$$

This means **18.18%** of the most recent value contributes to the EMA calculation.

By adjusting \( N \), the EMA can be tuned to react **faster** or **slower** to price movements.

---


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample stock price data
data = {'Date': pd.date_range(start='2024-01-01', periods=50, freq='D'),
        'Close': np.random.randint(100, 200, size=50)}
stock_data = pd.DataFrame(data)
stock_data.set_index('Date', inplace=True)

# Calculate Exponential Moving Average (EMA)
stock_data['EMA_10'] = stock_data['Close'].ewm(span=10, adjust=False).mean()
stock_data['EMA_20'] = stock_data['Close'].ewm(span=20, adjust=False).mean()

# Plotting the EMA
plt.figure(figsize=(12,6))
plt.plot(stock_data['Close'], label='Stock Price', color='blue', alpha=0.5)
plt.plot(stock_data['EMA_10'], label='10-Day EMA', color='orange', linewidth=2)
plt.plot(stock_data['EMA_20'], label='20-Day EMA', color='red', linewidth=2)

plt.title('Exponential Moving Average (EMA)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
