## Descriptive Statistics in Time Series Analysis
Descriptive statistics are summary measures used to describe the essential features of a dataset. In time series analysis, they help understand the central tendency, dispersion, and shape of the data distribution over time.

## üìå Key Descriptive Statistics for Time Series:

### 1Ô∏è‚É£ Measures of Central Tendency (Location)
These describe the **center** of the data.

- **Mean (Average)** ‚Üí The sum of all values divided by the count.  
$$
  \text{Mean} = \frac{\sum X_i}{N}
$$
  
- **Median** ‚Üí The middle value when data is sorted. It represents the central value where half the data points are smaller and half are larger.

- **Mode** ‚Üí The most frequently occurring value in the dataset.

---

### 2Ô∏è‚É£ Measures of Dispersion (Spread)
These show how much the data **varies**.

- **Variance** ‚Üí Measures the average squared deviation from the mean.  
  $$
  \text{Variance} = \frac{\sum (X_i - \bar{X})^2}{N}
  $$

- **Standard Deviation (SD)** ‚Üí The square root of variance, indicating how spread out the values are. It measures the **volatility** or **risk** in time series data.  
  $$
  \text{SD} = \sqrt{\text{Variance}}
  $$

- **Range** ‚Üí The difference between the maximum and minimum values in the dataset. It provides an understanding of the total spread of the data.

---

### 3Ô∏è‚É£ Measures of Shape
These describe the **distribution** of data.

- **Skewness** ‚Üí Measures the **asymmetry** of the data. A positive skew means the right tail is longer, and a negative skew means the left tail is longer.

- **Kurtosis** ‚Üí Measures the "tailedness" of the data. It indicates whether the data has **heavy** or **light tails**:
  - High kurtosis: Data has heavy tails (outliers are more likely).
  - Low kurtosis: Data has light tails (fewer outliers).


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample time series data
data = {'Date': pd.date_range(start='2024-01-01', periods=50, freq='D'),
        'Close': np.random.randint(100, 200, size=50)}
stock_data = pd.DataFrame(data)
stock_data.set_index('Date', inplace=True)

# Descriptive Statistics
print("Mean:", stock_data['Close'].mean())
print("Median:", stock_data['Close'].median())
print("Mode:", stock_data['Close'].mode().values)
print("Variance:", stock_data['Close'].var())
print("Standard Deviation:", stock_data['Close'].std())
print("Skewness:", stock_data['Close'].skew())
print("Kurtosis:", stock_data['Close'].kurt())

# Plot time series
plt.figure(figsize=(12,6))
plt.plot(stock_data['Close'], label='Stock Price', color='blue')
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

## Why Descriptive Statistics Matter in Time Series?

Descriptive statistics provide valuable insights into the characteristics of time series data, helping to understand its behavior and guide decision-making:

- ‚úÖ **Identifies trends & patterns**:  
   By looking at measures like the **mean**, you can identify if the data is trending upward or downward. For example, an increasing mean suggests an uptrend in the data over time.

- ‚úÖ **Measures volatility**:  
   The **standard deviation** shows the degree of variability or volatility in the data. A higher standard deviation indicates more fluctuations (higher risk), while a lower value indicates more stability.

- ‚úÖ **Detects skewness & outliers**:  
   Measures like **skewness** and **kurtosis** help identify if the data is skewed or has outliers. This is crucial for model selection, as certain models work better with symmetric distributions, while others handle skewed or heavy-tailed data more effectively.
